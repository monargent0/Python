```python
# FamilySize라는 컬럼을 생성 : SibSp + Parch + 1
train.head()
train['FamilySize'] = train.SibSp + train.Parch + 1
train.head(10)
```

```python
# C : 오스트리아 컬럼명 : Nationality_AT , S,Q : UK 컬럼명 : Nationality_UK
train['Nationality_AT'] = train.Embarked == 'C'
train['Nationality_UK'] = (train.Embarked == 'S') | (train.Embarked == 'Q')
train[['Embarked','Nationality_AT','Nationality_UK']].head()
```

```python
# Nationality
train.loc[train['Nationality_UK']==True, 'Nationality'] = 'England'
train.loc[train['Nationality_AT']==True, 'Nationality'] = 'Austria'
train.head()
```

```python
# Fare
train['Fare'].describe()
```

```python
# LowFare <15 MediumFare 15~30 HighFare 30>
train['LowFare'] = train['Fare'] < 15
train['MediumFare'] = (train['Fare'] >= 15)  & (train['Fare'] < 30)
train['HighFare'] = train['Fare'] >= 30


train[['Fare','LowFare','MediumFare','HighFare']].head(10)
```

```python
# FareType 컬럼

train.loc[train['LowFare']==True, 'FareType'] = 'Low'
train.loc[train['MediumFare']==True, 'FareType'] = 'Medium'
train.loc[train['HighFare']==True, 'FareType'] = 'High'

train[['Fare','FareType']]
```

```python
# fare에 train['Fare']데이터가 들어온다 
def check_fare_type(fare):
    if fare < 15:
        return "Low"
    elif fare >=15 and fare <30:
        return "Medium"
    else:
        return "High"
```

```python
# 함수로 생성하는 방식, 데이터가 많을 때는 함수를 사용하는게 더 빠르다 

train['FareType'] = train['Fare'].apply(check_fare_type)
train[['Fare','FareType']].head(10)
```

```python
# Age의 NaN을 Age의 평균값으로 대체하기 (평균나이는 소수점 1자리)
train['Age'].fillna(round(train['Age'].mean(),1) , inplace=True )

print(round(train['Age'].mean(),1))

train['Age']
```

```python
# 현재까지의 자료 저장하기
train.to_csv("../Data/new_train.csv")
```
