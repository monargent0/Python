```python
import nltk # natural language tool kit
from konlpy.corpus import kolaw
from konlpy.tag import Okt
import matplotlib
import matplotlib.pyplot as plt
from wordcloud import WordCloud

okt=Okt()

# 대한민국 헌법 불러오기
lawtxt = kolaw.open('constitution.txt').read()
lawtxt

nouns = okt.nouns(lawtxt) # 명사
print(nouns)
```

```python
#불용어 추가
stop_words = ['제','월','일','조','수','때','그','이','바','및','안','정']
nouns = [each_word for each_word in nouns if each_word not in stop_words]
nouns
```

```python
# 상위 50개 단어 확인
ko = nltk.Text(nouns)
most50 = ko.vocab().most_common(50)
```

```python
# 상위 50개 단어 리스트 
for i in most50:
    print(i[0],':', i[1] ,'회')
```
