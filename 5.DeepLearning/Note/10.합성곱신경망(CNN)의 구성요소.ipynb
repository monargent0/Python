분석으로 진로를 희망할때 기업에서 많이 물어보는것 CNN 합성곱, 머신러닝의 PCA(주성분분석)

### CNN(Convolution Neural Network)에서 본 심층신경망(다층 퍼셉트론)의 단점
- 입력값으로 고유 이미지 생김새 정보를 알 수 없다. 2차원 배열 자료를 1차원으로 변경하여 작업이 시작된다. // 여기서부터 글러먹었다.
- 심층신경망은 픽셀 하나 하나의 변화에 상당히 민감하다. 이미지 생김새를 사용할 수 없으므로 한두개의 픽셀이 모델 예측에 영향을 끼친다. // 낮 얼굴, 밤 얼굴을 구분하지 못한다.
- 픽셀 한 두개의 정보에도 민감하게 반응하기 위해 상당히 많은 변수를 모델 안에 가지고 있다. 이로인해 학습 시간이 오래 걸리고 잘못하면 과대적합이 발생한다.       

: 이것들을 해결하기 위해 나온것이 CNN이다

### 합성곱 신경망 (CNN : Convolution Neural Network)의 구성 요소
- 밀집층(Dense)       
  
1차원 합성곱 (특성을 찾는 것)       
- 합성곱 신경망은 노드의 가중치 갯수가 전체 갯수가 아니고 정하여 사용함 (3,5개 : 가중치가 3개 혹은 5개) (filter)
- 입력층을 아래로 한칸 이동하여 위에서 계산한 동일한 가중치와 절편을 사용한다.
- 밀집층보다 가중치의 갯수가 훨씩 적음.
- 10개의 입력이 있으면 출력은 8개만 만들어짐 (출력갯수 = 입력갯수 - 가중치 갯수 + 1)
- 3개씩 있는 것을 필터라고 하며 (기본값 3개) "필터의 커널은 3이다" 라고 한다.       

2차원 합성곱
- 이미지의 크기는 4X4
- 필터의 커널은 3X3이고 각 입력층에 곱해지는 Weight과 Bias는 동일하다.
- 발생하는 출력의 갯수는 4(2X2)개이다. 

특성맵
: 위에서 출력된 2X2가 특성맵이다.
- 특성맵을 활성화 출력이라고도 한다. 이유는 커널과 특성맵 사이에 활성화 함수(Relu)가 들어가기 때문이다.
- 그러나 CNN에서는 특성맵이라고 불리운다.
- 2차원으로 구성된 입력층으로 그대로 특성맵도 2차원으로 출력한다. 

여러개의 필터 
- 필터가 여러개 이므로 모든 가중치와 절편이 달라야 한다.
- 특성맵의 구성은 가로X세로X필터 의 갯수로 3차원이 생성됨.

* keras.laters.Conv2D(10, kernel_size = (3,3), activation = "relu") <- 10: 필터 갯수 , 커널 사이즈 , 활성함수 >

Padding 
- Same padding / Valid padding 
- 입력값(의 크기)과 동일한 특성맵의 크기를 구하는 방법으로 same Padding이 있으며, valid는 Padding이 없이 구하는 방법이다.

* keras.laters.Conv2D(10, kernel_size = (3,3), activation = "relu" , padding = "same")        

Stride  
- 필터의 이동을 한칸씩이 아닌 여러칸으로 이동할 경우, 이 이동하는 크기를 stride라고 한다. 
- 아주 특수한 경우를 빼고는 stride는 1을 초과하여 사용하지 않는다.
: 증명사진에는 사용하지 않음, 배경사진, 슬로우 영상(이미지가 거의 안 변할때) 같은 경우에 사용한다.

* keras.laters.Conv2D(10, kernel_size = (3,3), activation = "relu" , padding = "same", stride = 1)    

Pooling 층
- 특성맵의 다음단계로 차원축소를 하는 필터
- 최대풀링과 평균풀링의 2종류가 있다. (거의 최대풀링 사용)

최대풀링의 경우
- 풀링의 경우 계산하는 값의 픽셀이 겹치지 않는다.
- 풀링은 가중치가 없으며 풀링의 결과도 특성맵이라고 한다.
