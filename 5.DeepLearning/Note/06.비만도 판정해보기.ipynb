### Data 정규화

```python
# weight와 height에 대해서 정규화
csv['weight'] = csv['weight'] / csv['weight'].max()
csv['height'] = csv['height'] / csv['height'].max()
```

```python
csv.head()
```

```python
# label 바꿔주기
bclass = {'thin' : [1,0,0] , 'normal' : [0,1,0] , 'fat' : [0,0,1] }
y = []
for i , v in enumerate(csv['label']):
    y.append(bclass[v])

y
```

```python
# Training Set
X_train = csv[['height', 'weight']][0:15000]
y_train = y[0:15000]
```

```python
# Test Set
X_test = csv[['height', 'weight']][15001:20001]
y_test = y[15001:20001]
```

```python
print(len(X_train),len(X_test))
```

```python
print(len(y_train),len(y_test))
```

---
### Deep Learning 만들기

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(512 , activation='relu' , input_shape = (2,))) # Hidden layer
model.add(keras.layers.Dense(3 , activation='softmax')) # 출력층
```

```python
model.compile(
    optimizer = 'sgd', # 모든 옵션의 출발점은 경사하강법 , 옵션들은 거기서 발전한 것들. 
    metrics = ['accuracy'], # list 로 추가 옵션들도 넣을 수 있다.
    loss = 'categorical_crossentropy' # 정수가 아니라서 sparse_ 빠짐
)
```

--> 여기까지 딥러닝 환경을 다 만든것. 이제 데이터를 넣으면 된다.

```python
# 데이터 훈련하기
import numpy as np
model.fit( 
    np.array(X_train),
    np.array(y_train),
    epochs = 20
)
```

---
# 테스트 데이터로 평가하기

```python
score = model.evaluate( X_test , np.array(y_test)) 
# score = model.evaluate( np.array(X_test) , np.array(y_test))  # 에러나면 np씌워보고 하면 됨
print('loss : ' , score[0])
print('accuracy : ' , score[1])
```
