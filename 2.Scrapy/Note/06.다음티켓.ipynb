## 다음 영화 사이트
### https://movie.daum.net/ranking/boxoffice/weekly

- 탑건
- #mainContent > div > div.box_boxoffice > ol > li:nth-child(1) > div > div.thumb_cont > strong > a

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
url = "https://movie.daum.net/ranking/boxoffice/weekly"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")
```

```python
li_lists = soup.select("#mainContent > div > div.box_boxoffice > ol > li > div > div.thumb_cont > strong > a")
# soup.select("#mainContent .link_txt") # 짧은 버전
for list in li_lists:
    print(list.string)
```

```python
count = 0
titleLists = [] # 영화 제목 list
countLists = [] # 순위 list

for list in li_lists:
    count += 1
    print(count , list.string.strip())
    countLists.append(count)
    titleLists.append(list.string.strip())
```

```python
print(countLists , titleLists)
```

### 만든 리스트로 데이터 프레임 만들기

```python
# countlist 만으로 data frame 생성
df = pd.DataFrame(data = countLists, columns = [ '순위'])
df
```

```python
# 기존 Dataframe에 컬럼과 Lists 추가
df['영화제목'] = pd.Series(titleLists) # series는 R에서의 벡터 같은 역할
df
```

```python
# CSV로 저장하기
df.to_csv("../Data/DaumList.csv" , encoding="UTF-8", index = False)
```

```python
# 저장한 CSV 파일 확인하기
pd.read_csv("../Data/DaumList.csv")
```

### 네이버 영화 순위
- https://movie.naver.com/movie/sdb/rank/rmovie.naver
- 탑건
- #old_content > table > tbody > tr:nth-child(2) > td.title > div > a 

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd

url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")

movies = soup.select(".tit3 > a ")

# 데이터프레임 뼈대 만들기
df = pd.DataFrame(columns=['Rank', 'Titles'])

for i, movie in enumerate(movies):
    df.loc[i] = [i+1, movie.text.strip()]

# 데이터프레임 csv로 저장하기
df.to_csv('../Data/naverRanking.csv', encoding='utf-8', index=False)
```

```python
# 저장한 CSV 파일 확인하기
pd.read_csv("../Data/naverRanking.csv")
```
