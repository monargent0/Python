```python
# 라이브러리
from bs4 import BeautifulSoup
import urllib.request as req

# Site Address
# 기상청 RSS 사이트 주소
url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"

# urlOpen
res = req.urlopen(url)

# BeautifulSoup으로 분석하기 (BS한테 일 시키기)
soup = BeautifulSoup(res, "html.parser")
print(soup)
# print(soup.prettify()) # 들여쓰기 적용해서 보여줌
```

```python
# city 와 wf날씨 정보 가져오기
city = soup.find_all('city')
dic = {}

for i in city:
    name = i.string # 도시이름 
    wf = i.find_next('wf').string # city다음의 wf 찾아서 넣기

    if dic.get(wf) is None: # 키값이 없으면 새로 만들기
        dic[wf] = [name]
    else: # 키값이 있으면 데이터 추가하기
        dic[wf].append(name)

# print(dic)
# print(len(dic['흐리고 비']))

# 출력하기
for i in dic.keys():
    print("+",i)
    for j in range(len(dic[i])):
        print("| - ",dic[i][j])
```

```python
# 강사님 풀이
info = {}
for location in soup.find_all('location'):
    name = location.find('city').string
    weather = location.find('wf').string
    # print(name) # 내용 확인
    # print(weather) # 내용 확인

    if not(weather in info):
        info[weather] = []
    info[weather].append(name)
```

```python
# 각 지역의 날씨를 구분해서 출력하기
for weather in info.keys():
    print("+",weather)
    for city in info[weather]:
        print("| - ",city)

# 딕셔너리는 캘린더 만들때 자주 사용한다 
```

# JSON

```python
import urllib.request as req
import json

url = "https://zeushahn.github.io/Test/student.json"

res =req.urlopen(url)
soup = BeautifulSoup(res, "html.parser")

print(soup)
```

```python
# soup을 string으로 변형해서 json형태로 변형하기
items = json.loads(str(soup ))
print(items)
```

```python
# json의 object를 선택하기 
items2 = items['results']
print(items2)
```

```python
for item in items2 :
    print(item['code'],item['name'],item['dept'],item['phone'])
```
