```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

# Chrome Browser와 Chrome Driver Version 확인하기
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()), options = chrome_options)
driver.get('https://m.kyochon.com/store/store_list')

inputSearchtxt = driver.find_element(By.XPATH, '//*[@id="SEARCH_SHOP_NAM"]') # 검색에 값 넣기
inputSearchtxt.send_keys(" ") # 빈값 넣어주기
click_search = driver.find_element(By.XPATH, '//*[@id="btn_search"]').click() # 검색 버튼 누르기

# 셀레니움과 뷰티풀수프 합쳐서 사용하기
html = driver.page_source
print(html)

soup = BeautifulSoup(html, "html.parser")
```

```python
# 가맹점명 #store_list > li:nth-child(1) > h4
# 주소 지번 #store_list > li:nth-child(1) > div > p > span:nth-child(2)
import pandas as pd

names = soup.select("h4") 
addresses = soup.select("#store_list p span:nth-child(2)")

print(len(names))
print(len(addresses))


# 데이터프레임 뼈대 만들기
df = pd.DataFrame(columns=['가맹점명', '주소'])

for i, names in enumerate(names):
    df.loc[i] = [names.string.strip() , addresses[i].text.strip()[2:]]

print(df)

# 데이터프레임 csv로 저장하기
# df.to_csv('../Data/Kyochon_jungeun.csv', encoding='utf-8', index=False)
```

```python
### test
# 지역검색 선택
# //*[@id="app"]/div/div[3]/div[3]/div/div[1]/div[1]/div[1]/ul/li[3]/button
xpath = '//*[@id="app"]/div/div[3]/div[3]/div/div[1]/div[1]/div[1]/ul/li[3]/button'
click_city = driver.find_element(By.XPATH, xpath ).click() # 지역검색 버튼 누르기

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

# 자치구명 옵션 리스트
city = soup.select("div.form_list select:nth-child(1) option")
city.pop(0)
city
city_s = []
for c in city:
    city_s.append(c.string.strip())
city_s
```

```python
# test
name = soup.select(".tit strong")
adds = soup.select(".addr span:nth-child(1)")

gameng = []
address = []

for i in range(len(name)):
    gameng.append(name[i].string.strip())
    address.append(adds[i].string.strip())

len(gameng)
```

```python
# 화면에 그림을 그리는 시간이 필요하므로 이를 delay시키자!

import time

# # 데이터프레임 뼈대 만들기
# df = pd.DataFrame(columns=['가맹점명', '주소'])

gameng = []
address = []


for gu in city_s:
    # 시 선택 
    xpath = '//*[@id="app"]/div/div[3]/div[3]/div/div[1]/div[1]/div[2]/div[4]/div/select[1]'
    element = driver.find_element(By.XPATH, xpath) 
    element.send_keys(gu) 

    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    
    time.sleep(2) # wait 

    name = soup.select(".tit strong") # 가맹점 명
    adds = soup.select(".addr span:nth-child(1)") # 주소

    time.sleep(1) # wait 
    
    for i in range(len(name)):
        gameng.append(name[i].text.strip())
        address.append(adds[i].string.strip())
    
```

```python
# 111 + 30 + 21 + 24 + 16 + 15+9+6+109+9+12+9+15+7+27+26+1 = 447개 나와야 함
print(len(gameng),len(address))
```

```python
# 저장하기
# 데이터프레임 뼈대 만들기
df = pd.DataFrame(columns=['가맹점명', '주소'])

for i in range(len(gameng)):
    df.loc[i] = [gameng[i] , address[i]]

print(df)

# 데이터프레임 csv로 저장하기
df.to_csv('../Data/BurgerKing.csv', encoding='utf-8', index=False)
```

```python
# 화면에 그림을 그리는 시간이 필요하므로 이를 delay시키자!
import time

gameng = []
address = []

    # 시 선택 
elements = driver.find_elements(By.CSS_SELECTOR, ".area-name-select-left option")
for city in elements:
    
    city.click()

    time.sleep(1) # wait

    gu_s = driver.find_elements(By.CSS_SELECTOR, ".area-name-select-right option")
    
    if len(gu_s) > 1:

        for gu in gu_s:
            gu.click()
            time.sleep(1) # wait
            html = driver.page_source
            soup = BeautifulSoup(html, "html.parser")

            name = soup.select(".chain-title")
            adds = soup.select(".chain-address")

            for i in range(len(name)):
                gameng.append(name[i].string.strip())
                address.append(adds[i].string.strip())
    else :
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")

        name = soup.select(".chain-title")
        adds = soup.select(".chain-address")

        time.sleep(1) # wait
            
        for i in range(len(name)):
            gameng.append(name[i].string.strip())
            address.append(adds[i].string.strip())

```

```python
# 저장하기
# 데이터프레임 뼈대 만들기
df = pd.DataFrame(columns=['가맹점명', '주소'])

for i in range(len(gameng)):
    df.loc[i] = [gameng[i] , address[i]]

print(df)

# 데이터프레임 csv로 저장하기
df.to_csv('../Data/Alvolo.csv', encoding='utf-8', index=False)
```
